```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Data Analysis\Pandas file\Data Cleaning\File for Data Cleaning.xlsx")
df
```

```python
#Remove Duplicates. Duplicate drows will get removed.
#We assign the result to our df variable to save the changes. We can also use inplace as well to save the changes.

df = df.drop_duplicates()
df
```

```python
#Delete unused cols

df = df.drop(columns = "Not_Used_Col")
df
```

```python
#Remove extra dots, _ and / in LastName col. We use strip() function from str to do the same. 
#We take only the lastname col so we can apply the strip only to that col.
#We directly pass the values which we want to remove and strip will only remove these values if these are present @ start or end of the names.
#We save it to df["last_Name"], as if we save it to df, it will only save the last name data to df and will erase other cols from our df
#so be careful while saving the result of strip.

df["Last_Name"]=df["Last_Name"].str.strip(",._/")
df
```

# Change the phone number col to same format 123-456-7890 and also replace NAN and NA values to blank

```python
#1)USe regular expression [^a-zA-Z0-9], where we are taking values apart from ones mentioned in this and replacing it with 
#blank value

df["Phone_Number"] = df["Phone_Number"].str.replace("[^a-zA-Z0-9]","")
df
```

```python
#2)Now we want to change the format of the values to include hiphen in between. So we need to change the values to string, so
#we can put it in a loop using lambda expression/for loop and concat it with hiphen

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
#3)Here we use lambda function again to concat hiphen inbetween phone numbers

df["Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3]+"-"+x[3:6]+"-"+x[6:10])
df
```

```python
#4)Remove naan from phone number cols(if NA is present, use another replace() mtd to remove NA)

df["Phone_Number"] = df["Phone_Number"].str.replace("nan--","")
df
```

```python
#Splitting the address cols to 3 cols, 1)Street details 2)City 3)Pincode
#Expland attribute is used to expand the splitted cols into seperate cols.

df[["Street_Details","City","Pincodae"]]=df["Address"].str.split(",", expand=True)
df
```

```python
#In Paying Customer col, we are changing Yes->Y and No->N

df["Paying Customer"]= df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"]= df["Paying Customer"].str.replace("No","N")
df
```

```python
#In Don't Contact col, we are changing Yes->Y and No->N

df["Do_Not_Contact"]= df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"]= df["Do_Not_Contact"].str.replace("No","N")
df
```

```python
#Removing NAN(none) values from all cols

df=df.fillna("")
df
```

# Filtering the cols

```python
#We are removing the "Y" values from Do_Not_Contact cols as customers have said yes to not contact them. So we keep blank and 
#"N" values.
#df.loc[x], will provide values of all cols wrt index. We need only donot contact col values here, so we mention the col name in if statement.
#Then we drop the Y values in this col and save the same using inplace.

for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == "Y":
        df.drop(x,inplace=True)
        
df
```

```python
#Next we remove the customers who have not provided the phone number

for x in df.index:
    if df.loc[x,"Phone_Number"] == "":
        df.drop(x,inplace=True)
 
df
```

```python
#You can see the index values are not in order(0,8,11..). To reset it we use reset_index(), with drop as true(with out this, old index will be added as a col. Tis will drop the old index -> 0,8,11...)
df=df.reset_index(drop=True)
df
```
